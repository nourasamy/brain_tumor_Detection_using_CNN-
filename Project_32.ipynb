```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import BatchNormalization

from keras.preprocessing.image import ImageDataGenerator
from numpy.random import seed
seed(101)

import tensorflow  as tf
tf.random.set_seed(101)
```

```python
CNN_1=Sequential()

CNN_1.add(Convolution2D(32, 9,  input_shape=(32, 32, 3), activation='relu'))
CNN_1.add(MaxPooling2D(pool_size=(2, 2)))
#CNN_1.add(Dropout(0.25))
#CNN_1.add(BatchNormalization())
CNN_1.add(Convolution2D(16, 4, activation='relu'))
CNN_1.add(MaxPooling2D(pool_size=(2, 2)))
#CNN_1.add(Convolution2D(64, 4, activation='relu'))
#CNN_1.add(MaxPooling2D(pool_size=(2, 2)))
#CNN_1.add(Dropout(0.25))
#CNN_1.add(BatchNormalization())



CNN_1.add(Flatten())


CNN_1.add(Dense(units=128, activation='relu'))
#CNN_1.add(Dropout(0.25))
#CNN_1.add(BatchNormalization())


CNN_1.add(Dense(units=1, activation='sigmoid'))
CNN_1.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
train_datagen = ImageDataGenerator(rescale=1. / 255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)

test_datagen = ImageDataGenerator(rescale=1. / 255)
```

```python
training_set = train_datagen.flow_from_directory(r'C:\\Users\\Mazen\\Downloads\\brain_tumor_dataset\\brain_tumor_dataset\\training_set',
                                                 target_size=(32, 32),
                                                 batch_size=32,
                                                 class_mode='binary')

test_set = test_datagen.flow_from_directory(r'C:\\Users\\Mazen\\Downloads\\brain_tumor_dataset\\brain_tumor_dataset\\test_set',
                                            target_size=(32, 32),
                                            batch_size=32,
                                            class_mode='binary')
```

```python
CNN_1.fit_generator(training_set,
                         epochs=150,
                         validation_data=test_set )
CNN_1.save('tumormodel.h5');
print("finish")
```

```python
#Evaluate the model on training set
score = CNN_1.evaluate(training_set, verbose=0)
# Print training accuracy
print('\n', 'Training accuracy:', score[-1])
#Evaluate the model on test set
score = CNN_1.evaluate(test_set, verbose=0)
# Print test accuracy
print('\n', 'Test accuracy:', score[-1])
```
